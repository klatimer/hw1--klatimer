## 1.) Read the CSV file into a pandas dataframe

```python
# Download data if it is not present
import os
p = os.path.split(os.getcwd())
p = os.path.join(p[0], 'data')
if not os.path.exists(p):
    os.mkdir(p)
p = os.path.join(p, 'bikes.csv')
if not os.path.exists(p):
    from subprocess import call
    data_dir, _ = os.path.split(p)
    os.chdir(data_dir)
    call('curl -o bikes.csv https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD', shell=True)
else:
    pass
```

```python
# Load the CSV file into a pandas dataframe
import pandas as pd
df = pd.read_csv(p)
df.head()
```

## 2.) Add columns to the dataframe

```python
# Total east + west bicycle count
df['Fremont Bridge Both Sidewalks'] = df['Fremont Bridge East Sidewalk'] + df['Fremont Bridge West Sidewalk']
df.head()
```

```python
# Hour of the day and year
dt = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p')
list1 = []
list2 = []
for i in range(len(dt)):
    list1.append(dt[i].hour)
    list2.append(dt[i].year)
df['Hour'] = pd.Series(list1)
df['Year'] = pd.Series(list2)
df.head()
```
